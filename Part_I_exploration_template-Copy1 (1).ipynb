###  1. What are the variables that correlate more? 

```python
# plot matrix: sample 500 loan samples so that plots are clearer and they render faster
#To allow me to find correlation between variables, in form of any pattern that pops out.

Loan_samp = Loan_new.sample(n=500, replace = False)
print("Loan_samp.shape=",Loan_samp.shape)

numeric_sample= ['DebtToIncomeRatio', 'LoanOriginalAmount', 'StatedMonthlyIncome', 'BorrowerAPR','BorrowerRate','EstimatedLoss', 'EstimatedReturn']
g = sb.PairGrid(data = Loan_samp, vars = numeric_sample)
g = g.map_diag(plt.hist, bins = 20);
g.map_offdiag(plt.scatter)
```

Observation

2. The debt to income ratio, stated monthly income, loan original amount seems to be constant among other variables 
3. Borrowers rate has a positive correllation with Estimated Loss- would explore more

### 3. How does Borrower's APR and Rate correlate with Loan Term, and Homeowner and Employment Status?

```python
#To plot a graph of Borrower's rate and APR  in relation with Term , Homeowner and Employment Status.

Loan_samp = Loan_new.sample(n=2000, replace = False)

cat= ['Term','IsBorrowerHomeowner', 'EmploymentStatus']

def boxgrid(x, y, **kwargs):
    """ Quick hack for creating box plots with seaborn's PairGrid. """
    default_color = sb.color_palette()[0]
    sb.boxplot(x=x, y=y, color=default_color)

plt.figure(figsize = [15, 10])

g = sb.PairGrid(data = Loan_samp, y_vars = ['BorrowerAPR', 'BorrowerRate'], x_vars = cat,
                height = 3, aspect = 1.5)
g.map(boxgrid)
plt.show();
```

Observation

1. The most common term for the borrowers rate and rate is 36
2. Borrower's rate and APR is higher for non-home owner.
